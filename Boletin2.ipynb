#### Crea un programa que pida al usuario los datos de altura y base de un rectángulo. Tiene que tener al menos dos funciones en las que se les pase los datos de la altura y la base del rectángulo y devuelvan el área y el perímetro del mismo. Estos dos datos se tienen que mostrar al usuario por pantalla.


```python
def calcularAreaRectángulo(ancho, alto):
    return ancho * alto
def calcularPerímetroRectángulo(ancho, alto):
    return 2 * (ancho + alto)


ancho = float(input("Ingrese el ancho del rectángulo: "))
alto = float(input("Ingrese el alto del rectángulo: "))

área = calcularAreaRectángulo(ancho, alto)
perímetro = calcularPerímetroRectángulo(ancho, alto)

print(f"El área del rectángulo es: {área}")
print(f"El perímetro del rectángulo es: {perímetro}")
```

#### Crea un programa que pida al usuario dos años. Tiene que tener al menos una función donde se les pase ambos números y se calcule la cantidad de años bisiestos que hay entre esos dos años. Por pantalla se tienen que mostrar esta cantidad y que años son los bisiestos. Nota: el usuario puede meter los años en cualquier orden, es decir, no tiene por qué ser primero el mayor o el menor

```python
def esBisiesto(año1, año2):
    cantidadBisiestos = 0
    for año in range(año1, año2 + 1):
        if (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0):
            cantidadBisiestos += 1
    return cantidadBisiestos
    

año1 = int(input("Ingresa un año:"))
año2 = int(input("Ingresa otro año:"))

print(f"Cantidad de años bisiestos entre {año1} y {año2}: {esBisiesto(año1, año2)}")
```

#### Crea un programa que pida al usuario que introduzca una cantidad aleatoria de palabras, de una en una. Si el usuario no quiere introducir más palabras, que introduzca “FIN”. El programa tiene que devolver todas las palabras ordenadas alfabéticamente, teniendo en cuenta las mayúsculas y las minúsculas. Se debe usar al menos una función.

```python
def ordenarLista(lista):
    return sorted(lista)

palabras = []
palabra = ""
while palabra != "FIN":
    palabra = input("Ingresa una palabra (o 'FIN' para terminar): ")
    if palabra != "FIN":
        palabras.append(palabra)
palabras_ordenadas = ordenarLista(palabras)
print("Lista ordenada de palabras:", palabras_ordenadas)
```

```python
def ordenarLista(lista):
    return sorted(lista)

palabras = []
palabra = ""
while palabra != "FIN":
    palabra = input("Ingresa una palabra (o 'FIN' para terminar): ")

    if palabra not in palabras:

        if palabra != "FIN":
            palabras.append(palabra)

    else:
        print(f"La palabra {palabra} ya fue ingresada. Intenta con otra.")


palabras_ordenadas = ordenarLista(palabras)
print("Lista ordenada de palabras:", palabras_ordenadas)
```

#### Crea un programa que pida al usuario que introduzca una cantidad de horas, el coste €/hora y el porcentaje de impuestos. Se tiene que mostrar por pantalla el sueldo bruto, los impuestos que se pagarían y el sueldo neto. Para cada parámetro utilizar una función.

```python
def salario_bruto(cantidad_horas, tarifa_por_hora):
    return cantidad_horas * tarifa_por_hora

def salario_neto(cantidad_horas, tarifa_por_hora, porcentaje_impuestos):
    bruto = salario_bruto(cantidad_horas, tarifa_por_hora)
    impuestos = bruto * (porcentaje_impuestos / 100)
    return bruto - impuestos


cantidad_horas = int(input("Ingrese la cantidad de horas trabajadas en el mes: "))
tarifa_por_hora = float(input("Ingrese el coste por hora: "))
porcentaje_impuestos = float(input("Ingrese el porcentaje de impuestos a deducir: "))

bruto = salario_bruto(cantidad_horas, tarifa_por_hora)
neto = salario_neto(cantidad_horas, tarifa_por_hora, porcentaje_impuestos)

print(f"El salario bruto es: {bruto}")
print(f"El salario neto es: {neto}")
```

#### Crea un programa que pida al usuario que introduzca una cantidad de números aleatorios. Si el usuario no quiere introducir más números, que introduzca “-1”. El usuario no puede poder introducir datos que no sean números. Se mostrará por pantalla el promedio, el máximo y el mínimo de la lista de números introducidos. Se deben utilizar al menos 4 funciones.


```python

def maximo(lista):
    if not lista:
        return None
    max_val = lista[0]
    for num in lista:
        if num > max_val:
            max_val = num
    return max_val

def minimo(lista):
    if not lista:
        return None
    min_val = lista[0]
    for num in lista:
        if num < min_val:
            min_val = num
    return min_val

def promedio(lista):
    if not lista:
        return None
    return sum(lista) / len(lista)


def main():
    n = 0
    lista_numeros = []
    while n!= -1:
        n = int(input("Ingresa un número entero (o -1 para terminar): "))
        if n != -1:
            lista_numeros.append(n)
    print(f"Lista de números: {lista_numeros}")
    print(f"Número máximo: {maximo(lista_numeros)}")
    print(f"Número mínimo: {minimo(lista_numeros)}")
    print(f"Promedio: {promedio(lista_numeros)}")
    
if __name__ == "__main__":
    main()
```
